```python
#objective : to check whether data is stationary or not stationary
```

```python
#adfuller test
#null hypothesis: data is non stationary   p>=0.05 accept null hypothesis
#alternative    :  data is stationary      p<=0.05 reject null hypothesis                 data is stationary
```

```python
from statsmodels.tsa.stattools import adfuller
test=adfuller(df['#Passengers'])
if test[1]>=0.05:
    print('Accept null hypothesis data is non stationary')
else:
    print('Reject null hypothesis data is stationary')
```

```python
p_value=test[1]
p_value
```

```python
#Signal    =>Transform      =>Operations    =>Inverse transform           =>signal 
```

```python
#1. sqrt transform

df['sqrt']=np.sqrt(df['#Passengers'])
df.head()
```

```python
#check stationarity
test=adfuller(df['sqrt'])
if test[1]>=0.05:
    print('Accept null hypothesis data is non stationary')
else:
    print('Reject null hypothesis data is stationary')
```

```python
test[1]
```

```python
#2.cube root transform
df['cbrt']=np.cbrt(df['#Passengers'])
df.head()
```

```python
#check stationarity
test=adfuller(df['cbrt'])
if test[1]>=0.05:
    print('Accept null hypothesis data is non stationary')
else:
    print('Reject null hypothesis data is stationary')
```

```python
df['log']=np.log(df['#Passengers'])
df.head()
```

```python
#check stationarity
test=adfuller(df['log'])
if test[1]>=0.05:
    print('Accept null hypothesis data is non stationary')
else:
    print('Reject null hypothesis data is stationary')
```
