```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
sns.violinplot(data=df,x='Embarked',y='Fare')
```

```python
sns.boxplot(data=df,x='Embarked',y='Age')
```

```python
sns.catplot(data=df,x='Sex',hue='Survived',kind='count')
#barplot of sex vs count of male female who survived or not
```

```python
sns.violinplot(data=df,x='Sex',y='Age',hue='Survived',split='True')
#violinplot of sex vs age who surv or not
#Good for children.
#High for women in the age range 20-50.
#Less for men as the age increases.
```

```python
# Adding a column Family_Size 
df['Family_Size'] = 0
df['Family_Size'] = df['Parch']+df['SibSp'] 
 # Adding a column Alone 
df['Alone'] = 0
df.loc[df.Family_Size == 0, 'Alone'] = 1 #if family size is zero then set 1 in column alone
sns.factorplot(data=df,x='Family_Size',y='Survived',ylim=[0,1])  
sns.factorplot(data=df,x='Alone',y='Survived')
#If a passenger is alone, the survival rate is less.
#If the family size is greater than 5, chances of survival decrease considerably.
plt.show()
```

```python
df['Fair_range']=pd.qcut(df['Fare'],4)
sns.barplot(data=df,x='Fair_range',y='Survived')
#four bins of fair range vs survived or not 
#paid fair increases survival chances increases
```
